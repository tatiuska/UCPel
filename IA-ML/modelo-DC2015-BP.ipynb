```python
#Aplicação do ExtraTreesClassifier na base de dados de Doenças Cardíacas, levando em conta o atributo HighBP
#Carregando o conjunto de dados
import pandas as pd
arquivo = pd.read_csv('D:/UCPEL/IA-ML/HDI-2015.csv')
```

```python
arquivo.head()
```

```python
#Separando as variáveis entre preditoras e variável-alvo
y = arquivo['HighBP']
x = arquivo.drop('HighBP', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
arquivo.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier
for i in range(100):
    modelo = ExtraTreesClassifier()
    modelo.fit(x_treino, y_treino)
    
    resultado = modelo.score(x_teste, y_teste)
    print("Acurácia:", resultado)
```
